```python
def bone(x,y):
    """takes np arrays of x and y coords and computes a location in a 2x1 tile arranges in herringbone pattern"""
    yy = np.mod(np.floor(y)+np.floor(x),4)
    return ((yy<2)*np.mod(x,1) + (yy>=2)*np.mod(y,1), 
            (yy<2)*np.mod((y+np.floor(x))/2,1) + (yy==2)*np.mod((x+np.floor(y))/2,2) + (yy==3)*np.mod((x+np.floor(y))/2,2))
cbone = compose(gradientGB, bone)

fig, ax = pyplot.subplots(1, 4, figsize=(30,30))
showC(cbone, scale=2, centreX=2, centreY=2, axes=ax[0])
showC(cbone, scale=2.7, axes=ax[1])
showC(cbone, scale=5.1, axes=ax[2])
showC(cbone, scale=10., axes=ax[3])
```
